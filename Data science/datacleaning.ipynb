# Grouping the data by club and outputing players by age from the Bayern Munich club

```python
group_by_club = df.groupby('club')

group_by_club.get_group('Bayern Munich').sort_values(by='age', ascending=False)
```

```python
# Finding the average age of the players in Bayern Munich
group_by_club.get_group('Bayern Munich')['age'].mean()
```

```python
group_by_club.get_group('Bayern Munich')['age'].min()
```

# More comfortable way to find the average age of the players in Bayern Munich

```python
group_by_club.get_group('Bayern Munich').agg({"age":['min','max','mean'], 'price':['min','max','mean']})
```

```python
table = soup.find('table', class_ = 'standard')
```

```python
print(table)
```

to find only names of columns is th tags 
the data are in list

```python
world_title = table.find_all("th")
world_title
```

To input it into list

```python
world_table_titles = [title.text.strip() for title in world_title]
print(world_table_titles)
```

```python
import pandas as pd
```

from world_table_titles list we take only titles and inputing into dataframe

```python
df = pd.DataFrame(columns = world_table_titles)
df.drop(columns='Динамика численность населения в 1950-2022 годах[2]', inplace=True)
df
```

Searching information from the site except titles

```python
column_data = table.find_all('tr')
column_data
```

putting all this data into list

```python
ind_table_data = []
for row in column_data[2:]:
    row_data = row.find_all('td')
    world_table_data = [data.text.strip() for data in row_data]
    print(world_table_data)
    ind_table_data.append(world_table_data)
print(ind_table_data)

```

```python
lst = [] 
for i in ind_table_data:
    for j in i:
        lst.append(j)
lst
```

putting all this into dataframe 

```python
for row in column_data[2:]:
    length = len(df)
    df.loc[length] = lst
df
```

Deleting this NBSP

```python
df = df.replace('\xa0', '', regex=True)
df
```

indexing only needed part

```python
df.drop(df.index[1:], inplace=True)
df
```
